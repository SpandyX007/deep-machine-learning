```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
y
```

### The blue points are False and the green points are True

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap="winter",s=100);
```

```python
def step(z):
    if z>0:
        return 1
    else:
        return 0
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)   # inserting 1 at the 0th index i.e at the front
    weights=np.ones(X.shape[1])     # W0, W1, W2, ...
    lr=0.01     # learning rate
    epoch=1000
    for i in range(epoch):
        j=np.random.randint(0,100)      # select points randomly
        y_hat=step(np.dot(X[j],weights))    # dot product for X and weights
        weights=weights+lr*(y[j]-y_hat)*X[j]    # applying the formula
    return weights[0],weights[1:]
```

```python
intercept_,coeff_=perceptron(X,y)
```

```python
intercept_
```

```python
coeff_      # A and B in line Ax+By+c=0
```

```python
m=-(coeff_[0]/coeff_[1])
c=-(intercept_/coeff_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
